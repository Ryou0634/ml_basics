## 適応正則化分類

データ生成

```python
import numpy as np

np.random.seed(1)  # set the random seed for reproducibility
def data_generate(n=50):
    x = np.random.randn(n, 2)
    x[:n // 2, 0] -= 15
    x[n // 2:, 0] -= 5
    x[1:3, 0] += 10
    y = np.concatenate((np.ones(n // 2), -np.ones(n // 2)))
    index = np.random.permutation(np.arange(n))
    return x[index], y[index].astype('i')
```

```python
xs, ys = data_generate(200)
```

```python
%matplotlib inline
from scripts.visualizers import draw_scatter
draw_scatter(xs, (ys+1)//2)
```

適応正則化

```python
from models.adaptive_regularization_classifier import AdaptiveRegularizationClassifier

model = AdaptiveRegularizationClassifier()
model.fit(xs, ys, gamma=1)
draw_scatter(xs, (ys+1)//2, model)
```
